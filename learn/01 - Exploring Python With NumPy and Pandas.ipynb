## Loading a DataFrame from a file

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv'
df_students = pd.read_csv(url,delimiter=',',header='infer')
df_students.head()
```

## Handling missing values

```python
df_students.isnull()
```

```python
df_students.isnull().sum()
```

```python
df_students[df_students.isnull().any(axis=1)]
```

```python
df_students.StudyHours = df_students.StudyHours.fillna(df_students.StudyHours.mean())
df_students
```

```python
df_students= df_students.dropna(axis=0, how='any')
df_students
```

## Explore data in DataFrame

Membersihkan nilai yang hilang, tahap ini akan menjelajahi data di DataFrame.Dan membandingkan data covid dengan rata-rata.

```python
# Mendapatkan rata-rata jam belajar  menggunakan nama kolom sebagai indeks
mean_study = df_students['StudyHours'].mean()

# Menadapatkan rata-rata jam belajar menggunakan properti
mean_grade = df_students.Grade.mean()

# Menampilakan rata-rata hasil belajar dan nilai
print('Rata-rata jam belajar mingguan: {:.2f}\nNilai rata-rata: {:.2f}'.format(mean_study, mean_grade))
```

```python
# Mendapatkan rata-rata siswa yang belajar
df_students[df_students.StudyHours > mean_study]
```

```python
df_students[df_students.StudyHours > mean_study].Grade.mean()
```

```python
passes = pd.Series(df_students['Grade'] >= 60)
df_students = pd.concat([df_students, passes.rename("Pass")], axis=1)
df_students
```

```python
print(df_students.groupby(df_students.Pass).Name.count())
```

```python
print(df_students.groupby(df_students.Pass)['StudyHours', 'Grade'].mean())
```

```python
# Membuat suatu DataFrame dari nilai terbesar ke yang terkecil (descending)
df_students = df_students.sort_values('Grade', ascending=False)
df_students
```
