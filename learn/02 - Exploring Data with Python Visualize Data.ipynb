# Exploring Data with Python Visualize

Menerapkan analisis data dengan menggunakan statistik dasar dan memvisualisasikannya menggunakan grafik

```python
import pandas as pd

# Memuat data dalam file teks
url ='https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv'
df_students = pd.read_csv(url, delimiter=',', header='infer')

# Hapus baris dengan data yang hilang
df_students = df_students.dropna(axis=0, how='any')

# Hitung data yang lulus, dengan asumsi '60' adalah ketentuan nilai untuk kelulusan
passes = pd.Series(df_students['Grade'] >= 60)

# Simpan data yang yang diteruskan ke dataframe pandas
df_students = pd.concat([df_students, passes.rename("Lulus")], axis=1)

# Mencetak data frame
df_students
```

## Visualizing data with Matplotlib

```python
# Tampilkan plot dalam baris 
%matplotlib inline

import matplotlib.pyplot as plt

# Plot sumbu name x grade
plt.bar(x=df_students.Name, height=df_students.Grade)
plt.xticks(rotation=90)
# Tampilkan Plot
plt.show()
```

```python
# Membuat bilah nama x kelas
plt.bar(x=df_students.Name, height=df_students.Grade, color='Orange')

# Penyesuaian Grafik
plt.title('Student Grades')
plt.xlabel('Student')
plt.ylabel('Grade')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(rotation=90)

# Tampilkan Plot
plt.show()
```

```python
# Membuat Figur tampilan implisit
fig = plt.figure(figsize=(8,3))
fig = plt.figure()

# Menambahkan subplot 1
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot([1, 2, 3], [4, 5, 6])

# Menambahkan subplot 2
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot([1, 2, 3,], [4, 5, 6])

# Membuat bilah nama x kelas
plt.bar(x=df_students.Name, height=df_students.Grade, color='Orange')

# Menambahkan axes
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot([1, 2, 3], [4, 5, 6])

# Penyesuaian Grafik
plt.title('Student Grades')
plt.xlabel('Student')
plt.ylabel('Grade')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(rotation=90)

# Tampilkan Plot
plt.show()
```

```python
# Buat gambar untuk 2 subplot (1 baris, 2 kolom)
fig, ax = plt.subplots(1, 2, figsize = (10, 4))

# Buat plot bilah nama vs nilai pada sumbu pertama
ax[0].bar(x=df_students.Name, height=df_students.Grade, color='orange')
ax[0].set_title('Grades')
ax[0].set_xtiklabels(df_students.Name, rotation=90)

# Buat bagan pi chart jumlah kelulusan pada sumbu pertama
pass_counts=df_students['Pass'].value_counts()
ax[1].pie(pass_counts, labels=pass_counts)
ax[1].set_title('Passing Grades')
ax[1].legend(pass_counts.keys().tolist())

# Tambahkan judul ke gambar
fig.suptitle('Student Data')

# Tampilkan Gambar
plt.show()
```

## Statisical Analysis

#### Descriptive statistic and data Distribution 

```python
# Variable akan diperiksa
var_data = df_students['Grade']

# Buat Figur
fig = plt.figure(figsize=(10,4))

# Plot Histogram
plt.hist(var_data)

# Tambahan judul dan label
plt.title('Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Tampilkan figur
fig.show()
```

```python
# Membuat variable baru dan Menentukan nilai yang akan di analisis
var = df_students['Grade']

# Perhitungan Statistika
min_val= var.min()
max_val = var.max()
mean_val = var.mean()
med_val = var.median()
mod_val = var.mode()[0]

print('Minimum:{:.2f}\nMaen:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum{:.2f}\n'.format(min_val,
                                                                                      mean_val,
                                                                                      med_val,
                                                                                      mod_val,
                                                                                      max_val))
# Membuat figure
fig = plt.figure(figsize=(10,4))

# Membuat polt histogram
plt.hist(var)

# Menambahkan baris untuk digunakan statistika
plt.axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
plt.axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
plt.axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
plt.axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
plt.axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

# Menambahkan titles and labels
plt.title('Data Distribution')
plt.xlabel('value')
plt.ylabel('Frequnecy')

# Tampilkan Figure
fig.show()
```

```python
# Variable yang akan diperiksa 'Kelas'
var = df_students['Grade']

# Buat Figure
fig = plt.figure(figsize=(10,4))

# Plot Hisogram
plt.boxplot(var)

# Tambahkan judul dan label
plt.title('Data Distribution')

# Tampilkan Hasilnya
fig.show()
```

## Summary

As expected from the histogram of the sample, the density shows the characteristic "bell curve" of what statisticians call a normal distribution with the mean and mode at the center and symmetric tails.
