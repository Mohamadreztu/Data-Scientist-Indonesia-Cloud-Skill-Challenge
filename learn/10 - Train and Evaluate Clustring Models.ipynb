# Train and Evaluate Clustering Models

```python
import pandas as pd

# Memuat dataset untuk dlatih
url = 'https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/seeds.csv'
data = pd.read_csv(url)

# Menampilkan 10 sample acak observasi
features = data[data.columns[0:6]]
features.sample(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Meneormaliasasikan nilai fitur
scaled_features = MinMaxScaler().fit_transform(features[data.columns[0:6]])

# Mendapatkan 2 component prinsip
pca = PCA(n_components=2).fit(scaled_features)
features_2d = pca.transform(scaled_features)
features_2d[0:10]
```

```python
import matplotlib.pyplot as plt

%matplotlib inline 

plt.scatter(features_2d[:,0], features_2d[:,1])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Data')
plt.show()
```

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline

# Create 10 models with 1 to 10 clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    # Fit the data points
    kmeans.fit(features.values)
    # Get the WCSS (inertia) value
    wcss.append(kmeans.inertia_)
    
#Plot the WCSS values onto a line graph
plt.plot(range(1, 11), wcss)
plt.title('WCSS by Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
